### Assignment -3 :
Student Grade based on:

```python
marks = float(input("Enter the number (0-100): "))

if marks <0 or marks > 100:
    print("Invalid input!!")
    
elif marks >= 90 and marks <=100:
    print ("Grade: A+")

elif marks >= 80 and marks <=89:
    print ("Grade: A")

elif marks >= 70 and marks <=79:
    print ("Grade: B")    

elif marks >= 60 and marks <=69:
    print ("Grade: C")

else:
    print("Fail!!!")          
```

### Assignment -8 :
Leap Year Checker

```python
year =int(input("Enter the year: "))

if year % 400==0:
    print("Leap Year")

elif year % 4==0 and year % 100!=0:
    print("Leap Year")
    
else:
    print(" Not a Leap Year!!")    
```
